```python
import requests
from bs4 import BeautifulSoup
```

```python
infos = []

for i in range(1,11467):
    url = "https://movies.yahoo.com.tw/movieinfo_main/" + str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    
    info = {}
    
    if soup.find('div', class_='movie_intro_info_r'):
        for d in soup.find_all('div', id="content_l"):
            info['中文名稱'] = d.find('div', class_='movie_intro_info_r').find('h1').text.strip()
            info['英文名稱'] = d.find('div', class_='movie_intro_info_r').find('h3').text.strip()
            
            classification = []
            
            for j in d.find('div', class_='level_name_box').find_all('div', class_='level_name'):
                classification.append(j.find('a').text.strip())
                
            info['分類'] = classification
            info['劇情介紹'] = d.find('span', id ='story').text.strip()
            info['上映日期'] = d.find('span').text.strip()[5:]
            
            infos.append(info)
            print("No. " + str(i))
```

```python
import pandas as pd

df = pd.DataFrame(infos)
df.to_csv('movies.csv', encoding='utf_8_sig')
```
